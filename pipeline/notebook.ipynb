```python
import pandas as pd
```

# Read in data

```python
# Read a sample of the data as CSV
prefix = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/' # old data since only parquet now available on actual website
url = f'{prefix}/yellow_tripdata_2021-01.csv.gz'
```

```python
# dtypes to fix, datetime casting

dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates
)
```

# Add data to a Postgresql database

As there are more than 1 million rows, the data should be read in via chunking. Since it will take time, tdqm will monitor progress.

```python
 !uv add sqlalchemy psycopg2-binary
```

```python
# connect the new, empty database
from sqlalchemy import create_engine
engine = create_engine("postgresql://root:root@localhost:5432/ny_taxi") # postgresql, user root, password root, running on local  host, port is, database is called
```

```python
# start inserting data into database. Large database, so read in chunks

df_iter = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000
)
```

```python
from tqdm.auto import tqdm # progress monitoring
```

```python
for df_chunk in tqdm(df_iter):
    df_chunk.to_sql(name="yellow_taxi_data", con=engine, if_exists="append")
#to_sql(name="yellow_taxi_data", con=engine)
```
